```
from matplotlib import pyplot as plt
```

```
plt.plot([1,2,3],[4,5,1]) #x_values = [1,2,3] y_values = [4,5,6]==> coordinates (1,4),(2,5)..
plt.show()
```

```
#lets add title and variables to the graph
x = [1,2,3,4,5]
y = [1,4,9,16,25]
plt.plot(x,y, linewidth = 5, color = 'b')
plt.title("Plot")
plt.xlabel("x_values")
plt.ylabel("y_values")
plt.show()
```

```
plt.bar(x,y)
plt.title("Bar Graph") 
plt.xlabel("x_values")
plt.ylabel("y_values")
plt.show()
```

```
plt.scatter(x,y)
plt.title("Scatter Plot") #to compare two or three variables(ex finding correlation)
plt.xlabel("x_values")
plt.ylabel("y_values")
plt.show()
```

```
plt.hist(x,y)
plt.title("Histogram") #in bar plot we describe categorical variables(ex GDP growth), but in histogram quantitative variables like age group contributing to gdp growth
plt.xlabel("x_values")
plt.ylabel("y_values")
plt.show()
```

```
import numpy as np
def f(t):
  return np.sin(t)
```

```
t1 = np.arange(0,5,0.01)
t2 = np.arange(0,10,0.1)
#use subplots to show multiple plots. (abc)==> a=number of plots; b=no. of plots horizontally; c = present plot
plt.subplot(211)
plt.plot(t1,f(t1),'bo')
plt.subplot(212)
plt.plot(t2,f(t2))
```
